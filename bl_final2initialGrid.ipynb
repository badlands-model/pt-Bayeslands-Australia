```python
from pyBadlands.model import Model as badlandsModel
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.ndimage import filters
from scipy.spatial import distance
import random
```

```python
model = badlandsModel()
model.load_xml(str(1), 'Examples/etopo/etopo.xml', muted=True)
```

```python
final=np.loadtxt('Examples/etopo/data/final_elev.txt')
#Trim the boundary of the final array
final_trim=final[1:-1,1:-1]
elev_frame=pd.DataFrame({'X':model.recGrid.rectX,'Y':model.recGrid.rectY,'Z':final_trim.flatten()})
filename='Examples/etopo/data/evolved_demfile.csv'
#Save out the updated demfile
elev_frame.to_csv(filename,columns=['X', 'Y', 'Z'], sep=' ', index=False ,header=0,)
```

```python
#To look at it as a map, convert from Badlands format to a nice grid
xi=int(np.shape(model.recGrid.rectX)[0]/model.recGrid.nx)
yi=int(np.shape(model.recGrid.rectY)[0]/model.recGrid.ny)
#And put the demfile on a grid we can manipulate easily
elev=np.reshape(model.recGrid.rectZ,(xi,yi))
```

```python
plt.imshow(final_trim)
plt.show()
plt.imshow(elev)
plt.show()
```
